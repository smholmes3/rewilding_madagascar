```python
import pandas as pd
import numpy as np
import sklearn
import bioacoustics_model_zoo as bmz
```

```python
train_df = pd.read_csv(
    "data/train_labels_5s_mac_frommeta.csv",
    index_col=[0,1,2]
)

val_df = pd.read_csv(
    "data/val_labels_5s_mac.csv",
    index_col=[0,1,2]
)

test_df = pd.read_csv(
    "data/test_labels_5s_mac_frommeta.csv",
    index_col=[0,1,2]
)

classes = train_df.columns.tolist()
print(len(classes), classes[:5])
```

```python
m = bmz.BirdSetEfficientNetB1()
m.change_classes(classes)
```

```python
m.freeze_feature_extractor()

m.train(
    train_df,
    val_df,
    batch_size=32,
    num_workers=0,
    # if your OpenSoundscape version supports epochs as an argument, set a small number:
    epochs=5,
)
```

```python
m.save("data/birdset_effnetB1_finetuned_19cls.model", pickle=True)
```
